2D Tensor


```python
x = torch.tensor([
    [1, 2],
    [3, 4]
])

print(x)
print(f"x shape: {x.shape}")
```

```python
# 3D Tensor
x = torch.tensor([
    [
        [1,2],
        [3,4]
    ],
    [
        [5.5,6],
        [7,8]
    ]
])

print(x)
print(x.shape) # checks the shape of the tensor
print(x.dtype) # checks the data type of the tensor
print(x.device) # checks if the tensor is on CPU or GPU
print(torch.cuda.is_available()) # checks if GPU is available
```

##### Which numbers inside the model should change, and by how much? -> "Auto Grad" 🧠

Gradient = Tell which direct weights should change, and by how much so it can reduce the loss.

requires_grad =  tells PyTorch to track all operations on the tensor(number). 

backward() =  computes the gradient of the loss with respect to the parameters (or anything requiring gradients) using backpropagation.


    

<!-- #y -->

```python
# backward() -> backpropagation is the process of calculating the gradient of the loss function with respect to the model's parameters. It is a key step in training neural networks, allowing the model to learn from its errors and improve its performance.

weight = torch.tensor(2.0, requires_grad=True)


y = weight * 5
y.backward()
```

```python
import torch

x = torch.tensor(2.0, requires_grad=True)

y = x * 5

y.backward()

print(x.grad) # It tell which direct weights should change, and by how much so it can reduce the loss. 
```
