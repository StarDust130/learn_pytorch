# nn.Module  🧠

Every model in PyTorch looks like

``` class MyModel(nn.Module): ``` -> MyModel is a neural network.


````python
# Structure of Every PyTorch Model

import torch.nn as nn


class MyModel(nn.Module):  # Create a new neural network.

    def __init__(self):  # Runs once when the model is created.
        # Call the init method of the parent class nn.Module.
        super().__init__()

    # Defines the forward pass of the model. This is where the input data is passed through the layers of the model to produce an output.
    def forward(self, x):
        return x  # Return the output of the model.
````
