 Linear Layer => Linear transformation of the input data. It is also known as a fully connected layer or dense layer. It takes an input tensor and applies a linear transformation to produce an output tensor.

```python
import torch.nn as nn

layer = nn.Linear(2, 3)

print(layer.weight)
```

 ReLu => Ignore negative values, keep positive values.


```python
import torch
import torch.nn as nn

relu = nn.ReLU()

x = torch.tensor([-2, 5, -1, 8])

print(relu(x))
```

Dropout => randomly sets a fraction of input units to 0 at each update during training time, which helps prevent overfitting. 

- It is a regularization technique that improves the generalization of the model.

---

Embedding => A layer that maps discrete input tokens (like words or characters) to continuous vector representations. 

- It is commonly used in natural language processing tasks to convert categorical data into dense vectors that can be used as input to neural networks.

- Embedding is Word → Numbers

- Word ->  Vector (Tensor)

---

Conv2D => A convolutional layer that applies a 2D convolution operation to the input data. 

- It is commonly used in image processing tasks to extract features from images.
