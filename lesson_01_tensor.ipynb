Create Your First Tensor 🎉

```python
import torch
```

```python
x = torch.tensor([1, 2, 3])

print(x)
```

2D Tensor


```python
x = torch.tensor([
    [1, 2],
    [3, 4]
])

print(x)
print(f"x shape: {x.shape}")
```

```python
# 3D Tensor
x = torch.tensor([
    [
        [1,2],
        [3,4]
    ],
    [
        [5.5,6],
        [7,8]
    ]
])

print(x)
print(x.shape) # checks the shape of the tensor
print(x.dtype) # checks the data type of the tensor
print(x.device) # checks if the tensor is on CPU or GPU
print(torch.cuda.is_available()) # checks if GPU is available
```
