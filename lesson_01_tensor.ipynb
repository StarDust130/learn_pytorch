Create Your First Tensor 🎉

```python
import torch
```

```python
x = torch.tensor([1, 2, 3])

print(x)
```

2D Tensor


```python
x = torch.tensor([
    [1, 2],
    [3, 4]
])

print(x)
```

```python
# 3D Tensor
x = torch.tensor([
    [
        [1,2],
        [3,4]
    ],
    [
        [5,6],
        [7,8]
    ]
])

print(x)
```
